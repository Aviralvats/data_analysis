```python
import pandas as pd
```

```python
df=pd.read_csv('SalaryData.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(4)
```

```python
x=df[['Age']]
```

```python
y=df[['Salary']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
aviral=LinearRegression()
```

```python
aviral.fit(x,y)
```

```python
aviral.predict([[40]])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
a=aviral.coef_
```

```python
a
```

```python
b=aviral.intercept_
```

```python
b
```

```python
c=a*x+b
```

```python
plt.scatter(x,y)
plt.plot(x,c,color='red')
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
f'rmse value for given data is {root_mean_squared_error(y,c)}'
```

```python
f'r2 score for given data is {r2_score(y,c)}'
```
